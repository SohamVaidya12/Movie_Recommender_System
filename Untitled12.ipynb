```python
!pip install -q gradio pandas scikit-learn

import pandas as pd
import gradio as gr
import re
from urllib.parse import quote_plus
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# -------------------------------
# LOAD BOTH DATASETS
# -------------------------------
hollywood_movies = pd.read_csv("movie.csv")
bollywood_movies = pd.read_csv("bollywood_movies_only_2200.csv")

# -------------------------------
# CLEAN FUNCTION
# -------------------------------
def prepare_movies(df):
    df = df.copy()
    df["genres"] = df["genres"].fillna("")

    def extract_year(title):
        match = re.search(r"\((\d{4})\)", str(title))
        return match.group(1) if match else ""

    def clean_title(title):
        return re.sub(r"\(\d{4}\)", "", str(title)).strip()

    df["year"] = df["title"].apply(extract_year)
    df["clean_title"] = df["title"].apply(clean_title)

    df = df[df["genres"] != "(no genres listed)"].reset_index(drop=True)
    return df

hollywood_movies = prepare_movies(hollywood_movies)
bollywood_movies = prepare_movies(bollywood_movies)

# -------------------------------
# BUILD MODELS SEPARATELY
# -------------------------------
hollywood_tfidf = TfidfVectorizer(stop_words="english")
hollywood_matrix = hollywood_tfidf.fit_transform(hollywood_movies["genres"])

bollywood_tfidf = TfidfVectorizer(stop_words="english")
bollywood_matrix = bollywood_tfidf.fit_transform(bollywood_movies["genres"])

# -------------------------------
# COLLECT GENRES FROM BOTH FILES
# -------------------------------
def extract_genres(df):
    return {
        g.strip()
        for genre_list in df["genres"]
        for g in str(genre_list).split("|")
        if g.strip() and g.strip() != "(no genres listed)"
    }

hollywood_genres = extract_genres(hollywood_movies)
bollywood_genres = extract_genres(bollywood_movies)
all_genres = sorted(hollywood_genres.union(bollywood_genres))

# -------------------------------
# TRAILER LINK
# -------------------------------
def get_trailer_link(title, year, industry):
    query = f"{title} {year} official trailer {industry}"
    return f"https://www.youtube.com/results?search_query={quote_plus(query)}"

# -------------------------------
# MOVIE CARD
# -------------------------------
def make_movie_card(title, year, genres, match_pct, industry):
    trailer_url = get_trailer_link(title, year, industry)

    return f"""
    <div style="
        background:white;
        border-radius:18px;
        padding:16px;
        box-shadow:0 10px 25px rgba(0,0,0,0.12);
        margin-bottom:16px;
        border-left: 6px solid #ef4444;
    ">
        <h3 style="margin:0 0 8px 0; color:#111827;">{title}</h3>
        <p style="margin:6px 0;"><b>Year:</b> {year}</p>
        <p style="margin:6px 0;"><b>Genres:</b> {genres}</p>
        <p style="margin:6px 0;"><b>Match:</b> {match_pct}%</p>
        <p style="margin:6px 0;"><b>Industry:</b> {industry}</p>

        <a href="{trailer_url}" target="_blank" style="
            display:inline-block;
            margin-top:10px;
            background:#ef4444;
            color:white;
            padding:10px 16px;
            border-radius:10px;
            text-decoration:none;
            font-weight:600;
        ">
            ▶ Watch Trailer
        </a>
    </div>
    """

# -------------------------------
# MAIN RECOMMENDATION FUNCTION
# -------------------------------
def recommend_movies(industry, genre, num):
    if industry not in ["Hollywood", "Bollywood"]:
        return "<h3>Please select an industry.</h3>"

    if genre == "Select Genre":
        return "<h3>Please select a genre.</h3>"

    # Select dataset and model
    if industry == "Hollywood":
        movies = hollywood_movies
        tfidf = hollywood_tfidf
        matrix = hollywood_matrix
    else:
        movies = bollywood_movies
        tfidf = bollywood_tfidf
        matrix = bollywood_matrix

    # Transform selected genre
    user_vec = tfidf.transform([genre])
    sim_scores = cosine_similarity(user_vec, matrix).flatten()

    # Top matches
    top_indices = sim_scores.argsort()[::-1][:int(num)]

    # Build output cards
    html = """
    <div style="
        display:grid;
        grid-template-columns:repeat(auto-fit,minmax(260px,1fr));
        gap:18px;
        margin-top:20px;
    ">
    """

    for idx in top_indices:
        row = movies.iloc[idx]
        html += make_movie_card(
            title=row["clean_title"],
            year=row["year"],
            genres=row["genres"],
            match_pct=round(sim_scores[idx] * 100, 2),
            industry=industry
        )

    html += "</div>"
    return html

# -------------------------------
# OPTIONAL RANDOM GENRE
# -------------------------------
def surprise_genre():
    import random
    return random.choice(all_genres)

# -------------------------------
# GRADIO UI
# -------------------------------
custom_css = """
body {
    background: linear-gradient(to right, #141e30, #243b55);
}
.gradio-container {
    max-width: 1250px !important;
    margin: auto;
}
"""

with gr.Blocks(theme=gr.themes.Soft(), css=custom_css) as demo:
    gr.Markdown("""
    # 🍿 Movie Recommendation System
    ### Choose Bollywood or Hollywood, select a genre, and get smart movie recommendations
    """)

    with gr.Row():
        industry = gr.Radio(
            choices=["Hollywood", "Bollywood"],
            value="Hollywood",
            label="🎬 Select Industry"
        )

        genre = gr.Dropdown(
            choices=["Select Genre"] + all_genres,
            value="Select Genre",
            label="🎭 Choose Genre"
        )

        num = gr.Slider(
            minimum=1,
            maximum=10,
            step=1,
            precision=0,
            value=5,
            label="🎯 Number of Movies"
        )

    with gr.Row():
        recommend_btn = gr.Button("🚀 Recommend Movies", variant="primary")
        surprise_btn = gr.Button("🎲 Surprise Genre")



    output = gr.HTML(label="Recommended Movies")

    recommend_btn.click(
        fn=recommend_movies,
        inputs=[industry, genre, num],
        outputs=output
    )

    surprise_btn.click(
        fn=surprise_genre,
        outputs=genre
    )

demo.launch()
```
